```python
pattern_text = r'([\d]+) (To|-) ([\d]+)'
pattern = re.compile(pattern_text)
def findMean(value):
  match = re.match(pattern_text,value)
  min_value = match.group(1)
  max_value = match.group(3)
  return (float(min_value)+float(max_value))/2

for index,item in df1['Salary'].iteritems():
  if (type(item) == str):
    if " per Annum" in item:
      num = float(item.split(" per Annum")[0])
      df1.at[index,'Salary'] = str(num/12)

    elif "/Year" in item:
      num = float(item.split("/Year")[0])
      df1.at[index,'Salary'] = str(num/12)
    elif "K" in item:
      num = item.split("K")[0]
      df1.at[index,'Salary'] = num+"000"
    elif(re.match(pattern_text,item)):
      df1.at[index,'Salary']=findMean(item)
      updateErlist(str(index), df1['Salary'][index], 'Salary', item, findMean(item),'Salary Range provided', 'Data replaced by mean of the salary range') 
      
```
